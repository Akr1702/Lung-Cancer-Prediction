# LUNG CANCER PREDICTION


The primary goal is to predict whether a person has lung cancer based onvarious attributes related to their lifestyle, medical history, and symptoms.

Description of Features : 

❖ GENDER: Gender of the individual (M: Male, F: Female)

❖ AGE: Age of the individual

❖ SMOKING: Smoking status (YES = 2, NO = 1)

❖ YELLOW_FINGERS: Presence of yellow fingers, a symptom often associated with smoking (YES = 2, NO = 1)

❖ ANXIETY: Presence of anxiety (YES = 2, NO = 1)

❖ PEER_PRESSURE: Influence of peer pressure (YES = 2, NO = 1)

❖ CHRONIC_DISEASE: Presence of chronic disease (YES = 2, NO = 1)

❖ FATIGUE: Presence of fatigue (YES = 2, NO = 1)

❖ ALLERGY: Presence of allergy (YES = 2, NO = 1)

❖ WHEEZING: Presence of wheezing (YES = 2, NO = 1)

❖ ALCOHOL_CONSUMING: Alcohol consumption status (YES = 2, NO =1)

❖ COUGHING: Presence of coughing (YES = 2, NO = 1)

❖ SHORTNESS_OF_BREATH: Presence of shortness of breath (YES = 2,NO = 1)

❖ SWALLOWING_DIFFICULTY: Presence of difficulty in swallowing (YES= 2, NO = 1)

❖ CHEST_PAIN: Presence of chest pain (YES = 2, NO = 1)

❖ LUNG_CANCER: Presence of lung cancer (YES, NO)

## FEATURE SCALING

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
label_encoder = LabelEncoder()
data['GENDER'] = label_encoder.fit_transform(data['GENDER'])
data['LUNG_CANCER'] = label_encoder.fit_transform(data['LUNG_CANCER'])
```

```python
X = data.drop('LUNG_CANCER', axis=1)
Y = data['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=20)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled[:3]
```

```python
Y_train[:5]
```

# MODELING

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, Y_train)
```

```python
Y_pred = model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, Y_train)
```

```python
Y_pred_rf = rf_model.predict(X_test_scaled)

rf_accuracy = accuracy_score(Y_test, Y_pred_rf)
rf_accuracy
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, Y_train)
```

```python
Y_pred_dt = dt_model.predict(X_test_scaled)

dt_accuracy = accuracy_score(Y_test, Y_pred_dt)
dt_accuracy
```

## SVM

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(random_state=42)
svm_model.fit(X_train_scaled, Y_train)
```

```python
Y_pred_svm = svm_model.predict(X_test_scaled)

svm_accuracy = accuracy_score(Y_test, Y_pred_svm)
svm_accuracy
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, Y_train)
```

```python
Y_pred_nb = nb_model.predict(X_test_scaled)

nb_accuracy = accuracy_score(Y_test, Y_pred_nb)
nb_accuracy
```

## CONCLUSION :

### From above models SVM model have highest accuracy. So we use SVM Model for our Feature Selection

## FEATURE SELECTION

```python
from sklearn.feature_selection import RFE
```

```python
log_reg = SVC(kernel="linear", random_state=42)
rfe = RFE(estimator=log_reg, n_features_to_select=8)  
X_train_rfe = rfe.fit_transform(X_train_scaled, Y_train)
X_test_rfe = rfe.transform(X_test_scaled)
```

```python
log_reg.fit(X_train_rfe, Y_train)
Y_pred_log_reg = log_reg.predict(X_test_rfe)
```

```python
log_reg_accuracy = accuracy_score(Y_test, Y_pred_log_reg)
log_reg_report = classification_report(Y_test, Y_pred_log_reg)
log_reg_conf_matrix = confusion_matrix(Y_test, Y_pred_log_reg)
```

```python
print(f"SVM Accuracy with Feature Selection: {log_reg_accuracy}")
print("\nClassification Report:")
print(log_reg_report)
print("\nConfusion Matrix:")
print(log_reg_conf_matrix)
```

```python
selected_features = rfe.support_
feature_ranking = rfe.ranking_
```

```python
print("\nSelected Features (True = Selected, False = Not Selected):")
print(selected_features)
print("\nFeature Ranking (1 = Most Important):")
print(feature_ranking)
```

```python
feature_names = data.columns[:-1]
plt.figure(figsize=(10, 6))
plt.barh(range(len(feature_ranking)), feature_ranking, align='center', color='skyblue')
plt.yticks(np.arange(len(feature_names)), feature_names)
plt.xlabel('Feature Ranking (1 = Most Important)')
plt.ylabel('Feature Names')
plt.title('Feature Importance Ranking Using RFE with SVM')
plt.gca().invert_yaxis() 
plt.show()
```
